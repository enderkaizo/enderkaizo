```python
f = open("thongtin.csv","r",encoding = "utf8")
header = f.readline() 
data = f.readlines()

L_username = []
L_password = []
for i in data:
    i = i.strip()
    i = i.split(",")
    L_username.append(i[0])
    L_password.append(i[1])
    
from tkinter import *

sc = Tk()
sc.title("demogame")
def changedn1():
    fr1.pack_forget()
    fr2.pack()
def changeql1():
    fr2.pack_forget()
    fr1.pack()
def changedn2():
    fr3.pack_forget()
    fr2.pack()
def changedk1():
    fr1.pack_forget()
    fr3.pack()
def changedk2():
    fr2.pack_forget()
    fr3.pack()
    
def DN():
    username = var1fr2.get()
    password = var2fr2.get()
    if username == "" or password == "":
        lbtb_fr2.config(text = "Vui Lòng Nhập Đầy Đủ Thông Tin")
        return
    elif username not in L_username:
        lbtb_fr2.config(text = "Username Không Tồn Tại")
        return
    vt = L_username.index(username)
    if L_password[vt] != password:
        lbtb_fr2.config(text = "Sai Tên Đăng Nhập Hoặc Mật Khẩu")
        return
    fr2.pack_forget()
    fr4.pack()
    
def DK():
    signupname = var1fr3.get()
    signuppw = var2fr3.get()
    signupcpw = var3fr3.get()
    if signupname == "" or signuppw == "" or signupcpw == "":
        lbtb_fr3.config(text = "Vui Lòng Điền Đầy Đủ Thông Tin")
        return
    elif signupname in L_username:
        lbtb_fr3.config(text = "Username Đã Tồn Tại Vui Lòng Nhập Lại")
        return
    elif signuppw != signupcpw:
        lbtb_fr3.config(text = "Password Không Trùng Vui Lòng Nhập Lại")
        return
    f = open("thongtin.csv","a",encoding = "utf8")
    f.write(signupname+",")
    f.write(signuppw+"\n")
    f.close()
    L_username.append(signupname)
    L_password.append(signuppw)
    var1fr3.set("")
    var2fr3.set("")
    var3fr3.set("")
    fr3.forget_pack()
    fr2.pack()
    lbtb_fr2.config(text = "Mời Bạn Đăng Nhập Lại Để Vào Game")
    
    
fr1 = Frame(sc)

lb1_fr1 = Label(fr1,text = "Chào mừng bạn đã đến với trò chơi của tôi",font = ("Ariel",20),fg = "red",bg="black")
lb2_fr1 = Label(fr1,text = "Nếu đã có tài khoản bấm nút bên trái để đăng nhập còn nếu chưa có vui lòng bấm nút bên phải để đăng ký",font = ("Ariel",20),fg = "red",bg="black")
bt1_fr1 = Button(fr1,text = "Đăng Nhập",font = ("Ariel",20),padx=265,pady=50,fg = "blue",bg = "light green",command = changedn1)
bt2_fr1 = Button(fr1,text = "Đăng Ký",font = ("Ariel",20),padx = 265,pady=50,fg="cyan",bg = "red",command = changedk1)

lb1_fr1.grid(row = 1,column=1)
lb2_fr1.grid(row=2,column=1)
bt1_fr1.grid(row=4,column=1,sticky = W)
bt2_fr1.grid(row=4,column=1,sticky=E)
fr1.pack()

########
########

fr2 = Frame(sc)

lb3_fr2 = Label(fr2,text = "Vui Lòng Nhập Username và Password Để Đăng Nhập",font = ("Ariel",25))
lb1_fr2 = Label(fr2,text = "Username:",font = ("Ariel",20))
lb2_fr2 = Label(fr2,text = "Password:",font = ("Ariel",20))
var1fr2 = StringVar()
et1_fr2 = Entry(fr2,textvariable = var1fr2,font = ("Ariel",20),width=44)
var2fr2 = StringVar()
et2_fr2 = Entry(fr2,textvariable = var2fr2,font = ("Ariel",20),width = 44)
bt1_fr2 = Button(fr2,text = "Đăng Nhập",font = ("Ariel",25),command = DN,width= 20)
bt2_fr2 = Button(fr2,text = "Đăng Ký",font = ("Ariel",25),command=changedk2,width=20)
lbtb_fr2 = Label(fr2,text = "",font = ("Ariel",20))

lb3_fr2.grid(row=1,column=2)
lb1_fr2.grid(row=2,column=1)
et1_fr2.grid(row=2,column=2)
lb2_fr2.grid(row=3,column=1)
et2_fr2.grid(row=3,column=2)
bt1_fr2.grid(row=5,column=2,sticky=W)
lbtb_fr2.grid(row=4,column=2)
bt2_fr2.grid(row=5,column=2,sticky=E)

##########
##########

fr3 = Frame(sc)

lb1_fr3 = Label(fr3,text = "Vui Lòng Nhập Username Và Password Để Đăng Ký",font = ("Ariel",25))
lb2_fr3 = Label(fr3,text = "Username:",font = ("Ariel",20))
var1fr3 = StringVar()
et1_fr3 = Entry(fr3,textvariable = var1fr3,font = ("Ariel",20),width=44)
lb3_fr3 = Label(fr3,text = "Password:",font = ("Ariel",20))
var2fr3 = StringVar()
et2_fr3 = Entry(fr3,textvariable = var2fr3,font = ("Ariel",20),width=44)
lb4_fr3 = Label(fr3,text = "Password Confirm:",font = ("Ariel",20))
var3fr3 = StringVar()
et3_fr3 = Entry(fr3,textvariable = var3fr3,font = ("Ariel",20),width=44)
lbtb_fr3 = Label(fr3,text = "",font = ("Ariel",20))
bt1_fr3 = Button(fr3,text = "Đăng Ký",font = ("Ariel",25),width =20,command=DK)
bt2_fr3 = Button(fr3,text = "Đăng Nhập",font = ("Ariel",25),width=20,command=changedn2)

lb1_fr3.grid(row=1,column=2)
lb2_fr3.grid(row=2,column=1)
et1_fr3.grid(row=2,column=2)
lb3_fr3.grid(row=3,column=1)
et2_fr3.grid(row=3,column=2)
lb4_fr3.grid(row=4,column=1)
et3_fr3.grid(row=4,column=2)
lbtb_fr3.grid(row=5,column=2)
bt1_fr3.grid(row=6,column=2,sticky = W)
bt2_fr3.grid(row=6,column=2,sticky = E)

########
########

fr4 = Frame(sc)

bt1_fr4 = Button(fr4,text = "Bắt Đầu",font = ("Ariel",25),bg = "black",fg = "Red",width=11)
bt2_fr4 = Button(fr4,text = "Cài Đặt",font = ("Ariel",25),bg="black",fg="Gray",width=11)
bt3_fr4 = Button(fr4,text = "Đăng Xuất",font = ("Ariel",25),bg="black",fg="Gray",width=11)
bt4_fr4 = Button(fr4,text = "Thoát Game",font = ("Ariel",25),bg="black",fg="Gray",width=11)

bt1_fr4.grid(row=1,column=1,)
bt2_fr4.grid(row=2,column=1)
bt3_fr4.grid(row=3,column=1)
bt4_fr4.grid(row=4,column=1)





sc.mainloop()
```
